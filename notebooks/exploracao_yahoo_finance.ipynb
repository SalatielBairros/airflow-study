```python

def extrai_dados_diarios(codigo_acao: str, date: datetime): 
    caminho = f'/Users/salatiel/airflow/data/finances/extraction/{date.strftime("%Y%m%d")}_{codigo_acao}.csv'
    yfinance.Ticker(codigo_acao).history(
        interval = "1h",
        start = date.strftime("%Y-%m-%d"),
        end = (date + timedelta(days=1)).strftime("%Y-%m-%d"),
        prepost = True
    ).to_csv(caminho)
```

```python
extrai_dados_diarios("TOTS3.SA", datetime(2025, 12, 29))
```
